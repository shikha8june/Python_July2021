```python
import pandas as pd

url_data = (r'https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv')

data_csv = pd.read_csv(url_data)

data_csv.head()
```

```python
party_counts = pd.crosstab(data_csv['day'], data_csv['size'])
#type(party_counts)
party_counts
#We want to cross-tabulate between the day of the week and the size of the party. 
#In other words, we want to count how many parties of one were seated on Friday;
#how many parties of two; etc, 
#for each day of the week we have data (Friday, Saturday, Sunday, and Thursday). 
#To do this we will use crosstab.
```

```python
print(party_counts.sum(1))
```

```python
party_counts = party_counts.loc[:, 2:5] #Not so enlightening, because there is wide variation in the data. 
#Let's restrict our analysis to parties with a size between 2 and 5 (inclusive).
party_pcts = party_counts.div(party_counts.sum(1), axis=0)
print(party_pcts)
#We can "normalize" the daily data by dividing the values of a particular entry by the sum 
#of the values along that row. Notice that we use astype(float) to 
#make sure that no integer division problems are encountered, and 
#we specify axis=0 to say that we are normalizing along the day, not the party.
```

```python
party_pcts.plot.bar();
```

```python
party_pcts.plot(kind='bar', stacked=True)

plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5));
```
